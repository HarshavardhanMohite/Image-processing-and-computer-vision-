```python
import numpy as np
import cv2
```

```python
image=cv2.imread('/content/IMG_5134.JPG')
print(image.shape)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Resizing an image-cv2.resize()

```python
#used to change dimensions of an image
#it is resizing and image without preserving the ratio
height,width,_=image.shape

weight=500
height=500

image2=cv2.resize(image,(weight,height),interpolation=cv2.INTER_AREA)
print(image2.shape)
cv2_imshow(image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Resizing with preserving aspect ratio

```python


scale_factor=0.3
height,width,_=image.shape

new_height=int(height*scale_factor)
new_width=int(width*scale_factor)

image3=cv2.resize(image,(new_width,new_height),interpolation=cv2.INTER_AREA)
print(image3.shape)
cv2_imshow(image3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
