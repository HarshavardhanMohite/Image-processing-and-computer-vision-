```python
import  cv2
import numpy as np
import matplotlib.pyplot as plt
```

Display  grayscaleImage

```python
img = cv2.imread('/content/10.png')
from google.colab.patches import cv2_imshow
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)

print(img.shape)
#plt.imshow(img)
```

Mask

```python
SE = np.ones((15,15),np.uint8)
print(SE)
```

## Erosion

```python
print('erosion')
erosion = cv2.erode(img,SE,iterations=1)
cv2_imshow(erosion)
print(erosion.shape)
```

# Dilation

```python
dilation = cv2.dilate(img,SE,iterations=1)
#cv2_imshow(dilation)
plt.imshow(dilation, cmap='gray')
```

# opening

```python
# read the image
img = cv2.imread(r"/content/thresh.png", 0) #gray img
print("input img")
cv2_imshow(img)

# binari img
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
SE = np.ones((3, 3), np.uint8)

# opening the image
opening = cv2.morphologyEx(binr, cv2.MORPH_OPEN,SE, iterations=1)
print("output img")
cv2_imshow(opening)
```

# closing

```python
# read the image
img = cv2.imread(r"/content/thresh.png", 0) #gray img
print("input img")
cv2_imshow(img)

# binari img
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
SE = np.ones((3, 3), np.uint8)

# opening the image
opening = cv2.morphologyEx(binr, cv2.MORPH_CLOSE,SE, iterations=1)
print("output img")
cv2_imshow(opening)
```

# Gradient

```python
# read the image
img = cv2.imread(r"/content/th.jpeg", 0)
cv2_imshow(img)

# binarize the image
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]

# define the kernel
kernel = np.ones((3, 3), np.uint8)

# invert the image
invert = cv2.bitwise_not(binr)

# use morph gradient
morph_gradient = cv2.morphologyEx(invert,cv2.MORPH_GRADIENT,kernel)

# print the output
plt.imshow(morph_gradient, cmap='gray')
```
